```python
import numpy as np
import sklearn.linear_model as skmod
import matplotlib.pyplot as plt
```

```python
degree = 1
arr1 = [6,10,2,3,4,0,7,8,9,1] * degree
arr2 = [130,21, 43,76, 105, 3, 167, 162,91,15] * degree
```

```python
x1 = np.array(arr1).reshape(-1,1)
y = np.array(arr2).reshape(-1,1)

x2 = x1 * x1
x3 = x2 * x1


# print(x1)
# print(y)
# print(x2)
# print(x3)
```

```python

model_trained = skmod.LinearRegression().fit(x1, y)

w1 = model_trained.coef_[0][0]
b  = model_trained.intercept_[0]
r2 = model_trained.score(x1, y)

print(w1)
print(b)
print(r2)

# y' = 8.71*x + 37.75 
```

```python
arr_x1x2 = np.hstack([x1,x2]);
arr_x1x3 = np.hstack([x1,x2,x3]);

print(arr_x1x2)
print(arr_x1x3)
```

```python
model_trained2 = skmod.LinearRegression().fit(arr_x1x2, y)

w1 = model_trained2.coef_[0][0]
w2 = model_trained2.coef_[0][1]
b = model_trained2.intercept_[0]
r2 = model_trained2.score(arr_x1x2, y)

print(w1, w2, b)
print(r2)
# y' = 56.77 * x - 4.8 * x ^2 - 29.5
```

```python
model_trained3 = skmod.LinearRegression().fit(arr_x1x3, y)

w1 = model_trained3.coef_[0][0]
w2 = model_trained3.coef_[0][1]
w3 = model_trained3.coef_[0][2]
b = model_trained3.intercept_[0]

print(w1, w2, w3, b)
r2 = model_trained3.score(arr_x1x3, y)
print(r2)
# y' = 1.47 * x + 9.69 * x^2 -0.96 * x^3 + 5.27
```

```python
plt.scatter(x1, y)


plt.plot([0,10] ,[8.71*0 + 37.75, 8.71*10 + 37.75])
# plt.plot([0,10], model_trained.predict(np.array([[0],[10]])))- the same as  a equation
x_for_plot = np.arange(0, 10, 0.1) 
x2_for_plot = x_for_plot*x_for_plot
x3_for_plot = x2_for_plot*x_for_plot


x_final = np.hstack([x_for_plot.reshape(-1,1),x2_for_plot.reshape(-1,1) ])
y2_for_plot = model_trained2.predict(x_final)
plt.plot(x_for_plot, y2_for_plot)


x_final2 =  np.hstack([x_for_plot.reshape(-1,1),x2_for_plot.reshape(-1,1),x3_for_plot.reshape(-1,1) ])
y3_for_plot = model_trained3.predict(x_final2)
plt.plot(x_for_plot, y3_for_plot)
```
