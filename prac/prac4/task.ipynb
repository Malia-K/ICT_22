```python

import requests
import pandas as pd
import datetime
import matplotlib

def fromtimestamp(x):
    return datetime.datetime.fromtimestamp(x/1000.0)

resp = requests.get("https://api.polygon.io/v2/aggs/ticker/AAPL/range/1/minute/2022-08-01/2022-08-31?adjusted=true&sort=asc&limit=50000&apiKey=J76EPXiVXPVp1pZACagFjQ_G11ow1SKu")

resp_json = resp.json()

df = pd.DataFrame(resp_json['results'])


df['time'] = [fromtimestamp(df['t'][i]) for i in range(0, len(df)) ] 

df.plot(x = 'time', y = 'o')

# print(df)
```
