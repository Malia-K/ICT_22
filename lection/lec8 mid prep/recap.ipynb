```python
import requests
print(requests.__version__)
```

```python
resp = requests.get("https://api.thecatapi.com/v1/images/search")
print(resp.call)
```

```python
resp = requests.get("https://api.thecatapi.com/")
```

```python
print(resp.request.url)
print(resp.request)
print(resp.request.headers)
```

```python
print(resp)
print(resp.status_code)
print(resp.reason)
```

```python
print(resp.call) # use it more than text
print(resp.text)
print(resp.json())
```

```python
param = {'limit' : 10}
resp = requests.get("https://api.thecatapi.com/v1/images/search", params=param)

print(resp.request.url)
```

```python
resp = requests.get("https://www.kimep.kz/ru/")
print(resp.text)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(resp.content)
print(soup.find_all('a', href = 'https://www.kimep.kz/ru/blog/category/media/'))
```
